```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

sns.set_theme('notebook', palette='deep')
```

## Упражнение 1: Анализ изменений авиапассажиров

* Проанализировать изменение общего количества пассажиров за год (линейный график)
* Анализируйте распределение пассажиров по каждому месяцу года (гистограмма)

```python
data = sns.load_dataset('flights')
data.head()
```

```python
year_group_data = data.groupby(['year']).sum(numeric_only = True)

fig, ax = plt.subplots(figsize=(10,5))
ax = sns.lineplot(data=year_group_data, markers=True, legend=True)
ax.set_title('Количество клиентов в зависимости от года', fontsize = 16)
ax.set_xlabel('Year', fontsize = 14)
ax.set_ylabel('count', fontsize = 14)
ax.tick_params(axis='both', labelsize=10)
plt.show();
```

```python
fig, ax = plt.subplots(figsize = (10,5))

ax = sns.barplot(data=data, x='year', y ='passengers', hue='month')
ax.set_title('распределение пассажиров по месяцам в зависимости от года', fontsize = 16)
ax.set_xlabel('Year', fontsize = 14)
ax.set_ylabel('count', fontsize = 14)
ax.tick_params(axis='both', labelsize=10)
plt.ylim(0, 800)
plt.show();
```

```python
fig, ax = plt.subplots(figsize = (10,5))

ax = sns.barplot(data=data[data['year'] == 1960], x='month', y ='passengers')
ax.set_title('распределение пассажиров по месяцам за 1960 год', fontsize = 16)
ax.set_xlabel('Month', fontsize = 14)
ax.set_ylabel('count', fontsize = 14)
ax.tick_params(axis='both', labelsize=10)
plt.show();
```

## Упражнение 2: Анализ размера цветков ириса

* Соотношение размеров чашелистиков и лепестков (диаграмма рассеяния)
* Соотношение размеров между чашелистиками и лепестками разных видов ириса (классифицированная диаграмма рассеяния)
* Распределение размеров чашелистиков и лепестков разных видов цветов ириса (гистограмма или прямоугольная диаграмма)

```python
data = sns.load_dataset("iris")
data.head()
```

```python
data['sepal_size'] = data['sepal_length'] * data['sepal_width']
data['petal_size'] = data['petal_length'] * data['petal_width']

fig, ax = plt.subplots(figsize=(10,5))

ax = sns.scatterplot(data=data, x='sepal_size', y='petal_size', hue='species')
ax.set_title('Отношение размеров чашелистиков и лепестков', fontsize = 16)
ax.set_xlabel('sepal', fontsize = 14)
ax.set_ylabel('petal', fontsize = 14)

plt.show();
```

```python

ax.set_title('Распределение размеров чашелистиков и лепестков', fontsize = 16)
ax.set_xlabel('sepal', fontsize = 14)
ax.set_ylabel('petal', fontsize = 14)

[['sepal_size', 'petal_size', 'species']]
```

```python
fig, ax = plt.subplots(2, figsize=(8,10))

sns.histplot(data,x='sepal_size', hue='species', multiple='stack', ax=ax[0])
sns.histplot(data,x='petal_size', hue='species', multiple='stack', ax=ax[1])

ax[0].set_title('Распределение размеров чашелистиков и лепестков', fontsize = 16)
ax[0].set_xlabel('sepal size', fontsize = 14)
ax[0].set_ylabel('count', fontsize = 14)

ax[1].set_xlabel('petal size', fontsize = 14)
ax[1].set_ylabel('count', fontsize = 14)

plt.show();
```

```python
fig, ax = plt.subplots(2, figsize=(8,10))

sns.boxplot(data,x='sepal_size', y='species', ax=ax[0])
sns.boxplot(data,x='petal_size', y='species', ax=ax[1])

ax[0].set_title('Распределение размеров чашелистиков и лепестков', fontsize = 16)
ax[0].set_xlabel('sepal size', fontsize = 14)
ax[0].set_ylabel('count', fontsize = 14)

ax[1].set_xlabel('petal size', fontsize = 14)
ax[1].set_ylabel('count', fontsize = 14)

plt.show();
```

```python
fig, ax = plt.subplots(2, figsize=(8,10))

sns.violinplot(data,x='sepal_size', y='species', saturation = 1, ax=ax[0])
sns.violinplot(data,x='petal_size', y='species', saturation = 1, ax=ax[1])

ax[0].set_title('Распределение размеров чашелистиков и лепестков', fontsize = 16)
ax[0].set_xlabel('sepal size', fontsize = 14)
ax[0].set_ylabel('count', fontsize = 14)

ax[1].set_xlabel('petal size', fontsize = 14)
ax[1].set_ylabel('count', fontsize = 14)

plt.show();
```

## Упражнение 3: Анализ подсказок ресторана

* Связь между чаевыми и общим потреблением (диаграмма разброса)
* Клиенты-мужчины и покупатели-женщины, которые более щедры (диаграмма классифицированной коробки)
* Влияет ли курение на размер чаевых (диаграмма секретной коробки)
* В будние и выходные дни, когда клиенты дают более щедрые чаевые (диаграмма в рамке)
* За обедом или ужином, какой клиент даст больше чаевых (диаграмма классифицированной коробки)
* Влияет ли количество съеденных людей на щедрость (прямоугольная диаграмма категорий)
* Влияние сочетания пола и курения на щедрость (гистограмма группировки)

```python
data = sns.load_dataset("tips")
data.head()
```

```python
bill_mean = data['total_bill'].median()
tip_mean = data['tip'].median()

fig, ax = plt.subplots(figsize=(10,5))

sns.scatterplot(data=data, x='tip', y='total_bill')
ax.set_title('Связь между чаевыми и общим потреблением', fontsize = 16)
ax.set_xlabel('tip', fontsize = 14)
ax.set_ylabel('total bill', fontsize = 14)
ax.axline([tip_mean, bill_mean], [2*tip_mean, 2*bill_mean], c='red')

plt.show();
```

```python
fig, ax = plt.subplots(figsize=(10,5))

sns.boxplot(data=data, y='sex', x='tip')
ax.set_title('Связь между чаевыми и полом', fontsize = 16)

plt.show();
```

```python
data['tip%'] = data['tip']/(data['total_bill']/100)

fig, ax = plt.subplots(figsize=(10,5))

sns.boxplot(data=data, y='sex', x='tip%')
ax.set_title('Связь между %чаевых от чека и полом', fontsize = 16)

plt.show();
```

```python
fig, ax = plt.subplots(figsize=(10,5))

sns.boxplot(data=data, y='smoker', x='tip')
ax.set_title('Связь между чаевыми и курением', fontsize = 16)

plt.show();
```

```python
data['weekend'] = data['day'].apply(lambda x: 'Yes' if x in ['Sat', 'Sun'] else 'No')

fig, ax = plt.subplots(figsize=(10,5))

sns.boxplot(data=data, y='weekend', x='tip')
ax.set_title('Связь между чаевыми и выходными', fontsize = 16)

plt.show();
```

```python
fig, ax = plt.subplots(figsize=(10,5))

sns.boxplot(data=data, y='time', x='tip')
ax.set_title('Связь между временем приема пищи и чаевыми', fontsize = 16)

plt.show();
```

```python
fig, ax = plt.subplots(figsize=(10,5))

sns.boxplot(data=data, y='size', x='tip', orient='h')
ax.set_title('Связь между размером порции и чаевыми', fontsize = 16)

plt.show();
```

```python
mans_mean_tip = data[data['sex'] == 'Male']['tip'].mean()
womans_mean_tip = data[data['sex'] == 'Female']['tip'].mean()

fig, ax = plt.subplots(figsize = (5,5))

sns.barplot(data=data, x='sex', y='tip', hue='smoker', estimator='mean')
ax.set_title('Связь между полом, курением и чаевыми', fontsize = 16)

plt.show()
```
